### Data Analysis and Visualizations

```python
## Check How many percentage of bookings are canceled and plot a bar chart
canceled_perc = df['is_canceled'].value_counts(normalize=True)
print(canceled_perc)
plt.figure(figsize=(5,4))
plt.title('Reservation Status Count')
plt.bar(['Not Canceled','Canceled'],df['is_canceled'].value_counts(),edgecolor='k',width=0.7)
plt.show()
```

```python
## check which cancelation is more depending on the count
plt.figure(figsize = (8,4))
ax = sns.countplot(x ='hotel',hue = 'is_canceled', data = df,palette = 'Blues')
legend_labels,_ = ax. get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status in different hotels',size = 20)
plt.xlabel('hotel')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

* From the above graph we can say that the cancelation of 

```python
## Calculate percentage of cancelation in resort hotel
resort_hotel = df[df['hotel'] == 'Resort Hotel']
resort_hotel['is_canceled'].value_counts(normalize=True)
```

```python
## Calculate percentage of cancelation in city hotel
city_hotel = df[df['hotel'] == 'City Hotel']
city_hotel['is_canceled'].value_counts(normalize=True)
```

```python
## Check what is the effect of cancelation in city and resort hotel 
## (using groupby() on reservation status date and adr)

resort_hotel = resort_hotel.groupby('reservation_status_date')[['adr']].mean()
city_hotel = city_hotel.groupby('reservation_status_date')[['adr']].mean()
```

```python
## Plot Visualization for the avg. daily rate(adr) in both city and resort hotel
plt.figure(figsize = (20,8))
plt.title('Average Daily Rate in City and Resort Hotel', fontsize = 30)
plt.plot(resort_hotel.index, resort_hotel['adr'], label = 'Resort Hotel')
plt.plot(city_hotel.index, city_hotel['adr'], label = 'City Hotel')
plt.legend(fontsize = 20)
plt.show()
```

* From the above graph we can say that the price of city hotel is less than the resort hotel for some days (days can be holidays/weekends)because their is more flactuation in the data of resort hotel 
* This is the reason the cancelation in resort hotel is more in comparsion of city hotel.  

```python
## Lets check in which month reservation and cancelation is more in hotels

df['month'] = df['reservation_status_date'].dt.month     ## extract month from the reservation_status _Date (bcoz we hv no separate column for month)
plt.figure(figsize = (16,8))
ax = sns.countplot(x = 'month' , hue = 'is_canceled',data =df ,palette='bright')
legend_labels,_ = ax. get_legend_handles_labels()
ax.legend(bbox_to_anchor=(1,1))
plt.title('Reservation Status Per Month',size = 20)
plt.xlabel('month')
plt.ylabel('number of reservation')
plt.legend(['not canceled','canceled'])
plt.show()
```

* from the above graph we can say that :
* Cancelation is higher in month - Jan
* Cancelation is low in month  - Aug/Sept
* Reservation is high in month - Aug
* Reservation is low in the month - Dec

#### Case : (Reason of cancelation/reservation)
* Is price is less in aug month ? (Reservation is more)
* Is price is high in aug month ? (cancelation is more)

```python
## Check the effect of price on cancelation and plot adr for each month
plt.figure(figsize = (15,8))
plt.title('ADR per month', fontsize = 30)
sns.barplot(x ='month', y ='adr', data = df[df['is_canceled'] == 1].groupby(['month'])[['adr']].sum().reset_index())
plt.show()
```

* From the above graph we can say that price in jan month is high so that cancelation is high in this month.


```python
## Check in which country cancelation is more(top 10 country)
cancelled_data = df[df['is_canceled'] == 1]
top_10_country = cancelled_data['country'].value_counts()[:10]
plt.figure(figsize = (8,8))
plt.title('Top 10 countries with reservation canceled')
plt.pie(top_10_country, autopct = '%.2f', labels = top_10_country.index)
plt.show()
```

* From the above graph we can say that the cancelation in portugle is more

```python
## lets check price of reservation of either cancelled/ not cancelled
cancelled_df_adr = cancelled_data.groupby('reservation_status_date')[['adr']].mean()
cancelled_df_adr.reset_index(inplace = True)
cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

not_cancelled_data = df[df['is_canceled'] == 0]
not_cancelled_df_adr = not_cancelled_data.groupby('reservation_status_date')[['adr']].mean()
not_cancelled_df_adr.reset_index(inplace = True)
not_cancelled_df_adr.sort_values('reservation_status_date', inplace = True)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```

* From the above graph we say that data of reservation not cancelled is inconsistent from (2016 - 2017)

```python
## lets filter data from 2016 - 2017 from more clarity about reservatin cancelation/ not cancelation

cancelled_df_adr= cancelled_df_adr[(cancelled_df_adr['reservation_status_date']>'2016') & (cancelled_df_adr['reservation_status_date']<'2017-09')]
not_cancelled_df_adr = not_cancelled_df_adr[(not_cancelled_df_adr['reservation_status_date']>'2016') & (not_cancelled_df_adr['reservation_status_date']<'2017-09')]
```

```python
## plot data of (2016 to 2017)

plt.figure(figsize = (20,6))
plt.title('Average Daily Rate',fontsize = 30)
plt.plot(not_cancelled_df_adr['reservation_status_date'], not_cancelled_df_adr['adr'], label = 'not cancelled')
plt.plot(cancelled_df_adr['reservation_status_date'], cancelled_df_adr['adr'], label = 'cancelled')
plt.legend()
plt.show()
```

* From the above plot we can say that the cancelled data of (avg. daily rate) is more than the not cancelled data.
* Hence price effect the cancellation.
